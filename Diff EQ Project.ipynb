## II. Using the method of integrating factor, find the exact solution y(t) of the intial value problem.

Initial Value Problem: dy/dt = 1 + t - 5y, y(0) = 1

Integrating factor: u(t) = e^5t

Put equation in correct form: d/dt((e^5t)*y) = (e^5t)*t + (e^5t)

Solve: y = (5*(e^5t)*t + 4 * (e^5t) + 21) / 25 * (e^5t)


## III. Plot the exact solution found in II and your numerical solutions on the same axes. Compare these curves and draw a conclusion.


```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
```

```python

t = np.linspace(0, 5)
y = ((5 * (np.exp(5*t)) * t) + 4 * (np.exp(5*t)) + 21) / (25 * (np.exp(5*t)))
plt.plot(t, y, 'b-', linewidth=2.0)

approximateT = np.array([0])
approximateY = np.array([1])

t0 = 0
y0 = 1

stepSize = 0.1
numSteps = 50

for j in range(0, numSteps):
    m = 1 + t0 - 5*y0
    y1 = (1 - 5 * stepSize) * y0 + stepSize + stepSize * t0
    t1 = t0 + stepSize
    approximateT = np.append(approximateT, [t1])
    approximateY = np.append(approximateY, [y1])
    t0 = t1
    y0 = y1
    
    
plt.plot(approximateT, approximateY, 'r--', linewidth=2.0)

plt.show()
```
